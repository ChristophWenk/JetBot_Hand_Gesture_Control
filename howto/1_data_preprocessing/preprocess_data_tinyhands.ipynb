```python
import os
import cv2
import matplotlib.pyplot as plt
import random
import csv
```

```python
# ToDO: Set your path to the images folder and the csv file containing the bounding boxes
IMAGES_DIR = "C:/datasets/tinyhands/images300x300/"
LABELS_PATH = "C:/datasets/tinyhands/labels_all.csv"
```

```python
def read_csv(filename):
    """ Reads an existing csv file.
    Parameters:
        filename (str): eg. filename.csv
    Returns:
        list: all rows in a list
    """
    with open(filename, 'r') as f:
        reader = csv.reader(f)
        return list(reader)
```

```python
def test_labels():
    csv_list = read_csv(LABELS_PATH)[1:] # read labels without header
    index = random.randint(0, len(csv_list)-1)

    csv_record = csv_list[index]

    img_path = IMAGES_DIR+ csv_record[0]
    print(img_path)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    xmin, xmax, ymin, ymax = int(csv_record[1]), int(csv_record[2]), int(csv_record[3]), int(csv_record[4])

    rectangle = plt.Rectangle((xmin,ymin), xmax-xmin, ymax-ymin, facecolor='none', edgecolor="orange")
    plt.gca().add_patch(rectangle)
    plt.imshow(img)
```

```python
test_labels()
```
